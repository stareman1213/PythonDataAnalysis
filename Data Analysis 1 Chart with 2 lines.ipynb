```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from pylab import figure, show, legend, ylabel , xlabel
from matplotlib.ticker import FuncFormatter, MaxNLocator
```

```python
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
```

```python
df = pd.read_csv('.//data//NPQuery_data.csv' , low_memory=False)
df.columns
```

```python
fig1 = figure(figsize=(10,5))

# and the first axes using subplot populated with data 
plt = fig1.add_subplot(111)
line1, = plt.plot(df['avgTime'], 'o-',label="avgTime" , lw=4 )
ylabel(u"平均回應時間",family='Microsoft YaHei', size=20)
plt.legend(handles=[line1], loc=1)
# plt.set_xlabel(df['date'])

plt = fig1.add_subplot(111, sharex=plt, frameon=False)
line2, =plt.plot(df['count'], 'or-' ,label="count" ,lw=4)
plt.yaxis.tick_right()
plt.yaxis.set_label_position("right")

ylabel(u"筆數",family='Microsoft YaHei', size=20)
xlabel(u'日期',family='Microsoft YaHei', size=20)
# new_ticks = np.linspace(1, len(df['date']), len(df['date']))
# plt.set_xticks(new_ticks)
# plt.set_xticklabels(df['date'],rotation = 90 , size=15)
plt.legend(handles=[line2], loc=2)

xs = range(len(df['date']))
labels = list(df['date'])
def format_fn(tick_val, tick_pos):
    if int(tick_val) in xs:
        return labels[int(tick_val)]
    else:
        return ''

plt.xaxis.set_major_formatter(FuncFormatter(format_fn))
plt.xaxis.set_major_locator(MaxNLocator(integer=False))
# legend([line1, line2], ["1", "2"])
show()
```
